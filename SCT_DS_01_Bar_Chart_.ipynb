```python

import matplotlib.pyplot as plt

gender = ['Male', 'Female']
count = [60, 40]

plt.bar(gender, count)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution of Population')
plt.show()
```
